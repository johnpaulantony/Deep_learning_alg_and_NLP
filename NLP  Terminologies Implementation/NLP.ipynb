```python
# Tokenization ( Splitting text into sentences - words)

import nltk
from nltk.tokenize import sent_tokenize,word_tokenize

nltk.download('punkt_tab') # breaking down large text into sentence and words without requiring manually labled training data


text = "Natural Language Processing (NLP) is fun. It helps machines understand human language."

# word Token

word_tokens=word_tokenize(text)
print(word_tokens)

# sentence Token

sentence_tokens = sent_tokenize(text)
print(sentence_tokens)
```

```python
# REmoving stopwords

import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('punkt_tab')
nltk.download('stopwords')

text = "Natural Language Processing (NLP) is fun. It helps machines understand human language."

#Tokenize text
words=word_tokenize(text)

# GEt English stopwords
stop_words=set(stopwords.words('english'))

# Remove stopwords
filtered_words = [word for word in words if word.lower() not in stop_words]

print("Original Text:")
print(text)
print("\nFiltered Text:")
print(filtered_words)
```

```python
# Stemming - reduce words to their base form - rule based suffix stripping

from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

text = "Natural Language Processing (NLP) is fun. It helps machines understand human language."

ps=PorterStemmer()

# Toenize text into words

words=word_tokenize(text)
print(words)

# Apple stemping'
stemmed_words=[ps.stem(word) for word in words]

print("orginal words", words)

print("Stemmed words",stemmed_words)
```

```python
# Lemmatization ( similar to stemming but returns maningful words)

from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')

lemmatized_words=WordNetLemmatizer()

lemmatized_words=[lemmatized_words.lemmatize(word)for word in words]

print("orginal words", words)

print("Lemmatized words",lemmatized_words)
```

```python
# Parts-of-speech (POS ) Tagging (Noun,Verb,adjective)

from nltk import pos_tag

nltk.download('averaged_perceptron_tagger_eng')

# Tokenize text into words


pos_tags=pos_tag(words)
print("POS Tags",pos_tags)
```

```python
# Named Entity REcognition (NER) - identifying proper nouns, locations, organizations, persons etc

import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag, ne_chunk

nltk.download('maxent_ne_chunker_tab')
nltk.download('words')


# sample_text
text = "Barack Obama was the 44th President of the United States and lived in Washington."


# Tokenization & POS tagging

words = word_tokenize(text)
pos_tags = pos_tag(words)

# Named Entity Recognition
named_entities=ne_chunk(pos_tags)

print("Named Entities:")
print(named_entities)
```

```python

# BoW - Bag of words representation ( converting text into numerical format)
from sklearn.feature_extraction.text import CountVectorizer

# Sample text data
documents =  ["I love Python Programming.",
              "Python is great for AI and ML.",
              "AI and ML are the future."
              ]

vectorizer=CountVectorizer()

X=vectorizer.fit_transform(documents)

print(vectorizer.get_feature_names_out())

print(X.toarray())

```

Calculate TF-IDF

            TF-IDF = TF * IDF

            TF - TErm frequency - How often a word appears in a doc
            IDF - Inverse Doc freq - Measures how unique word is across documents

documents =  ["I love Python Programming.",
              "Python is great for AI and ML.",
              "AI and ML are the future."
              ]


    TF = Number of times a word appears in a doc/ total words in doc

    eg: python   1/4 = 0.25

    IDF = log(total no of documents / number of documents containing the word )

    eg: python
             log(3/2) = log(1.5) =0.176


        TF-IDF = TF * IDF  = 0.25 * 0.176 = 0.044

```python
N-Gram
      continous sequence of words - N words in a given text.

  default - unigram - "I am going to Bangalore" - ['I', 'am', 'going', 'to','Bangalore'] - Individual words

  Bigram - Two -word combinations ['I am','am going','going to','to Bangalore']

  Trigram - Three word combinations ['I am going','am going to','going to Bangalore']


  Bigram & trigram - gives more context and meaning



```

```python
# Bigram adnd trigram analysis ( finding word pairs and triples)

from nltk.util import ngrams
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt_tab')
#from collections import Counter

text = "Barack Obama was the 44th President of the United States and lived in Washington."

bigrams=list(ngrams(word_tokenize(text),2))
print("Bigrams:",bigrams)

trigrams=list(ngrams(word_tokenize(text),3))
print("Trigrams:",trigrams)

```
