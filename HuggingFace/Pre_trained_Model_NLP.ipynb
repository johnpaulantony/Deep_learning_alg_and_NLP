```python
import gradio as gr
from transformers import pipeline
summarizer = pipeline("summarization",model="facebook/bart-large-cnn")

def summarize_content(Article):
  summary=summarizer(Article,max_length=50,min_length=30,do_sample=False)
  return summary[0]['summary_text']

iface=gr.Interface(
    fn=summarize_content,
    inputs=gr.Textbox(lines=10,placeholder="enter the content"),
    outputs=gr.Textbox(label="summary"),
    title="Summarize data",
    description="Enter the content to be summarized"
)

iface.launch()
```

```python
# text translation, summarization, sentiment analysis

import gradio as gr
from transformers import pipeline

# Load Hugging Face Pipelines
translation_pipeline=pipeline("translation_en_to_fr")
summarization_pipeline = pipeline("summarization",model='facebook/bart-large-cnn')
sentiment_pipeline = pipeline("sentiment-analysis")

# Define the functions for each feature
def translate_text(text):
    return translation_pipeline(text)[0]['translation_text']
def summarize_content(text):
    return summarization_pipeline(text, max_length=100, min_length=30, do_sample=False)[0]['summary_text']
def analyze_sentiment(text):
    sentiment = sentiment_pipeline(text)[0]
    return sentiment['label']

# Gradio interface layout
with gr.Blocks() as demo:
    gr.Markdown("# Text Summarization, Translation, and Sentiment Analysis")

    with gr.Tab("Translation"):
      input_text=gr.Textbox(label="Enter text to translate")
      output_text=gr.Textbox(label="Translated text")
      translate_button=gr.Button("Translate")
      translate_button.click(fn=translate_text, inputs=input_text, outputs=output_text)

    with gr.Tab("Summarization"):
      input_text=gr.Textbox(label="Enter text to summarize")
      output_text=gr.Textbox(label="Summary")
      summarize_button=gr.Button("Summarize")
      summarize_button.click(fn=summarize_content, inputs=input_text, outputs=output_text)

      with gr.Tab("Sentiment Analysis"):
        input_text=gr.Textbox(label="Enter text to analyze sentiment")
        output_text=gr.Textbox(label="Sentiment")
        sentiment_button=gr.Button("Analyze Sentiment")
        sentiment_button.click(fn=analyze_sentiment, inputs=input_text, outputs=output_text)

#launch the interface
demo.launch()
```

```python
# text translation, summarization, sentiment analysis

!pip install gradio transformers torch
import gradio as gr
from transformers import pipeline
import torch

# Check if CUDA is available and set device accordingly
if torch.cuda.is_available():
    device = torch.device("cuda")
    print("CUDA is available. Using GPU.")
else:
    device = torch.device("cpu")
    print("CUDA is not available. Using CPU.")

# Load Hugging Face Pipelines
# Specify a smaller model for translation to avoid GPU issues
translation_pipeline = pipeline("translation_en_to_fr", model="t5-small", device=device) #using a smaller model
summarization_pipeline = pipeline("summarization", model='facebook/bart-large-cnn', device=device) # specify device
sentiment_pipeline = pipeline("sentiment-analysis", device=device) # specify device

# Define the functions for each feature
def translate_text(text):
    return translation_pipeline(text)[0]['translation_text']

def summarize_content(text):
    return summarization_pipeline(text, max_length=100, min_length=30, do_sample=False)[0]['summary_text']

def analyze_sentiment(text):
    sentiment = sentiment_pipeline(text)[0]
    return sentiment['label']

# Gradio interface layout
with gr.Blocks() as demo:
    gr.Markdown("# Text Summarization, Translation, and Sentiment Analysis")

    with gr.Tab("Translation"):
        input_text = gr.Textbox(label="Enter text to translate")
        output_text = gr.Textbox(label="Translated text")
        translate_button = gr.Button("Translate")
        translate_button.click(fn=translate_text, inputs=input_text, outputs=output_text)

    with gr.Tab("Summarization"):
        input_text = gr.Textbox(label="Enter text to summarize")
        output_text = gr.Textbox(label="Summary")
        summarize_button = gr.Button("Summarize")
        summarize_button.click(fn=summarize_content, inputs=input_text, outputs=output_text)

    with gr.Tab("Sentiment Analysis"):
        input_text = gr.Textbox(label="Enter text to analyze sentiment")
        output_text = gr.Textbox(label="Sentiment")
        sentiment_button = gr.Button("Analyze Sentiment")
        sentiment_button.click(fn=analyze_sentiment, inputs=input_text, outputs=output_text)

# Launch the interface
demo.launch()
```
