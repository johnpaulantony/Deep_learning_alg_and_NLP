Step 2: Set Up the Hugging Face Model

```python
from diffusers import StableDiffusionPipeline
import torch

# Load the stable diffusion pipeline
pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4", torch_dtype=torch.float16)
pipe = pipe.to("cuda")  # If you're using a GPU


import gradio as gr

# Define the prediction function
def generate_image_from_text(prompt):
    # Generate the image
    image = pipe(prompt).images[0]
    return image

# Set up the Gradio interface
interface = gr.Interface(fn=generate_image_from_text,
                         inputs="text",
                         outputs="image",
                         title="Stable Diffusion Text-to-Image Generator",
                         description="Enter a description, and the Stable Diffusion model will generate an image based on it.")

if __name__ == "__main__":
    interface.launch()
```
